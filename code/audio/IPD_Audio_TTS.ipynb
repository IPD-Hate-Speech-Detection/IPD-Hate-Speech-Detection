```python
import pandas as pd
df = pd.read_parquet("hf://datasets/ucberkeley-dlab/measuring-hate-speech/measuring-hate-speech.parquet")
```

```python
df = df[['text', 'hate_speech_score']]
```

```python
grouped_df = df.groupby('hate')
subset = grouped_df.sample(n=500, random_state=42)
subset.head()
```

```python
import os
import time
from gtts import gTTS
from tqdm import tqdm

def generate_tts_audio(df, output_dir='tts_audio', batch_size=50, sleep_interval=2):
    os.makedirs(output_dir, exist_ok=True)
    audio_paths = [None] * len(df)

    for batch_start in tqdm(range(0, len(df), batch_size), desc="Processing Batches"):
        batch_end = min(batch_start + batch_size, len(df))
        batch_df = df.iloc[batch_start:batch_end]

        for index, row in batch_df.iterrows():
            try:
                if pd.isna(row['text']) or not row['text'].strip():
                    continue

                tts = gTTS(text=str(row['text']), lang='en')
                filename = f'audio_{index}.mp3'
                filepath = os.path.join(output_dir, filename)
                tts.save(filepath)
                audio_paths[index] = filepath

            except Exception as e:
                print(f"Error processing row {index}: {e}")

        time.sleep(sleep_interval)

    df['audio'] = audio_paths

    return df
```

```python
df_final = generate_tts_audio(subset.reset_index(), batch_size=50, sleep_interval=2, output_dir=audio_df_path)
```

```python
df_final = generate_tts_audio(subset.tail(431).reset_index(), batch_size=50, sleep_interval=2, output_dir=audio_df_path)
```

## Merging Datasets

```python
df_1 = pd.read_csv('dataset_with_audio.csv')
df_2 = pd.read_csv('dataset_with_audio_2.csv')
```

```python
merged_df = pd.concat([df_1, df_2], ignore_index=True)
merged_df.reset_index(drop=True, inplace=True)
```

```python
merged_df.dropna(inplace=True)
```

```python
final_df = merged_df.drop_duplicates(subset=['index'], keep='first')
```

```python
org_df = pd.read_parquet("hf://datasets/ucberkeley-dlab/measuring-hate-speech/measuring-hate-speech.parquet")
```

```python
org_df = org_df[['text', 'hate_speech_score']]
org_df['hate'] = org_df['hate_speech_score'].apply(lambda x: 1 if x >= -1 else 0)
org_df.drop(columns=['hate_speech_score'], inplace=True)
org_df = org_df.reset_index()
```

```python
new_subset = org_df[org_df['index'].isin(final_df['index'])]
```

```python
column_name = 'index'
common_values = list(set(new_subset[column_name]).intersection(set(final_df[column_name])))
print(f"Number of common values: {len(common_values)}")
print("Common values:", common_values)
```

```python
common_column = 'index'
new_column = 'audio'
output = pd.merge(new_subset, final_df[[common_column, new_column]], on=common_column, how='left')
output.head()
```
